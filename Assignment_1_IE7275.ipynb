```python
import numpy as np

def sentropy(ts):
  entropy_value = 0
  #iterating each value in ts
  for i in ts:
      entropy_value = entropy_value + (i *np.log2(i)) #calculating entropy as per formula
  return (-1)*entropy_value

#creating an empty array for input
x = []
#taking input of array size
l = int(input("Enter the input array size :"))
#reading the elements of array
for i in range(l):
  a = float(input("Enter value :"))
  x.append(a)

x = np.array(x)

#function call for sentropy
print(sentropy(x))
```

ANSWER 2.3

```python
# subsetting the data to hold the columns: State, credit, WIC, WICcash and SNAP
fm_sub = fmdf.loc[:, ['State', 'Credit', 'WIC', 'WICcash', 'SNAP']]

# to replace all occurences in all columns
fm_sub = fm_sub.replace(to_replace = "Y", value = 1)
fm_sub = fm_sub.replace(to_replace = "N", value = 0)

# get Count of each payment type
fm_sub = pd.pivot_table(fm_sub, index = ['State'],
                               values = ['Credit', 'WIC', 'WICcash', 'SNAP'],
                               aggfunc = 'sum').reset_index()

# converting table to wide format
fm_sub = fm_sub.melt(id_vars = ['State'],
                                   var_name = 'Payment System',
                                   value_name = '#Farmers Market').sort_values(by = 'State')
# printing the fm_sub
print(fm_sub.to_string(index = False))
```
